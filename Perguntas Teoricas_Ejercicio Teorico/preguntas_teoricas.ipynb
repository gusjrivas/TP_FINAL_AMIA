#Pregunta 1

Para simplificar la expresión a la utilizada en LDA, donde todas las clases comparten la misma matriz de covarianza debemos seguirestos pasos:

a) Asumimos Covarianza Común:

Σ𝑗=Σ∀𝑗

b) Eliminar Dependencia de 𝑗 en la Covarianza:

Dado que la covarianza es la misma para todas las clases, el determinante

∣Σ∣ es constante para todas las clases y puede ser absorbido en la constante
𝐶. Entonces, podemos escribir:

$$
-\frac{1}{2}\log |\Sigma_j| = -\frac{1}{2}\log |\Sigma|
$$

Este término no depende de 𝑗 y se puede considerar parte de la constante
𝐶.

c) Simplificar el término restante:

$$
- \frac{1}{2} (x-\mu_j)^T \Sigma_j^{-1} (x- \mu_j) = - \frac{1}{2} (x-\mu_j)^T \Sigma^{-1} (x- \mu_j)
$$

d) Agrupar todos los términos que no dependen de 𝑗 en una nueva constante

$$
\log{f_j(x)} = - \frac{1}{2} (x-\mu_j)^T \Sigma^{-1} (x- \mu_j) + C'
$$

e) Expandimos el producto:
$$
(x - \mu_{j})^{T} \Sigma^{-1} (x - \mu_{j}) = x^{T} \Sigma^{-1} x - x^{T} \Sigma^{-1} \mu_{j} - \mu_{j}^{T} \Sigma^{-1} x + \mu_{j}^{T} \Sigma^{-1} \mu_{j}
$$

f) Agrupemos los términos que tienen x y los términos constantes:

$$
\log{f_{j}(x)} = -\frac{1}{2} (x^{T} \Sigma^{-1} x - 2x^{T} \Sigma^{-1} \mu_{j} + \mu_{j}^{T} \Sigma^{-1} \mu_{j}) + C'
$$

g) Podemos reorganizar los términos para obtener la forma deseada:

$$
\log{f_{j}(x)} = x^{T} \Sigma^{-1} \mu_{j} - \frac{1}{2} x^{T} \Sigma^{-1} x + C' - \frac{1}{2} \mu_{j}^{T} \Sigma^{-1} \mu_{j}
$$
$$
\log{f_{j}(x)} = \mu_{j}^{T} \Sigma^{-1} x - \frac{1}{2} x^{T} \Sigma^{-1} x + C' - \frac{1}{2} \mu_{j}^{T} \Sigma^{-1} \mu_{j}
$$

h) Podemos combinar $$C' - \frac{1}{2} \mu_{j}^{T} \Sigma^{-1} \mu_{j}
$$ en una nueva constante C'':

$$
\log{f_{j}(x)} = \mu_{j}^{T} \Sigma^{-1} x - \frac{1}{2} x^{T} \Sigma^{-1} x + C''
$$

i) Se reorganizan los términos para obtener la forma final:
$$
\log{f_{j}(x)} = \mu_{j}^{T} \Sigma^{-1} x - \frac{1}{2} \mu_{j}^{T} \Sigma^{-1} \mu_{j} + C''
$$

$$
\log{f_{j}(x)} = \mu_{j}^{T} \Sigma^{-1} x - \frac{1}{2} (\mu_{j}^{T} \Sigma^{-1} \mu_{j}) + C''
$$

$$
\log{f_{j}(x)} = \mu_{j}^{T} \Sigma^{-1} x - \frac{1}{2} \mu_{j}^{T} \Sigma^{-1} \mu_{j} + C''
$$

$$
\log{f_{j}(x)} = \mu_{j}^{T} \Sigma^{-1} (x - \frac{1}{2} \mu_{j}) + C''
$$

#Pregunta 2

a) En cuanto a LDA sabemos que se asume que cada clase comparte la misma matriz de covarianza, Σ. Por ende, la función a posteriori resulta de la siguiente manera:

$$
P(G|_{X=x} = j)  \propto f_j(x) \cdot \pi_j
$$

$$
\log[P(G|_{X=x} = j)]  \propto \log[f_j(x) \cdot \pi_j]
$$

$$
\log[P(G|_{X=x} = j)]  \propto \log[f_j(x)] + \log \pi_j
$$

$$
\log[P(G|_{X=x} = j)]  \propto x^T \Sigma^{-1} \mu_j - \frac{1}{2} \mu_j^T \Sigma^{-1} \mu_j + \log \pi_j
$$

Esta función es **lineal** en x porque el término con más peso de x es $$x^T \Sigma^{-1} \mu_k$$ es una función lineal de xx.

b) Por otro lado, QDA, se permite que cada clase tenga su propia matriz de covarianza $$Σ_j$$
Por ende, la función a posteriori resulta de la siguiente manera:

$$
P(G|_{X=x} = j)  \propto f_j(x) \cdot \pi_j
$$

$$
\log[P(G|_{X=x} = j)]  \propto \log[f_j(x) \cdot \pi_j]
$$

$$
\log[P(G|_{X=x} = j)]  \propto \log[f_j(x)] + \log \pi_j
$$

$$
\log[P(G|_{X=x} = j)]  \propto - \frac{1}{2} \log \left| \Sigma_j \right| - \frac{1}{2} (x - \mu_j)^T \Sigma_j^{-1} (x - \mu_j) + \log \pi_j
$$

Esta función es **cuadrática** en x debido al término $$\frac{1}{2} (x - \mu_j)^T \Sigma_j^{-1} (x - \mu_j)$$ es una forma cuadrática de x.

#Pregunta 3

Recordando que para una matriz A, el determinante de su inversa es igual al inverso de su determnante:

   $$
    \det(A^{-1}) = \frac{1}{\det(A)}
   $$

Por otro lado, usando la propiedad de logaritmos

   $$
   \log\left(\frac{1}{a}\right) = -\log(a)
   $$

Ahora, apliquemos estas propiedades a la expresión por definición de QDA:

$$
\log{f_j(x)} = -\frac{1}{2}\log |\Sigma_j| - \frac{1}{2} (x-\mu_j)^T \Sigma_j^{-1} (x- \mu_j) + C
$$

Resulta openado lo siguiente:

$$
\log{f_j(x)} = -\frac{1}{2}\log |\frac{1}{\Sigma_j^{-1}}| - \frac{1}{2} (x-\mu_j)^T \Sigma_j^{-1} (x- \mu_j) + C
$$

$$
\log{f_j(x)} = -\frac{1}{2}[-\log |\Sigma_j^{-1}|] - \frac{1}{2} (x-\mu_j)^T \Sigma_j^{-1} (x- \mu_j) + C
$$

$$
\log{f_j(x)} = \frac{1}{2}\log |\Sigma_j^{-1}| - \frac{1}{2} (x-\mu_j)^T \Sigma_j^{-1} (x- \mu_j) + C
$$

Que es la expresión utilizada en el código. De esta manera, hemos demostrado que ambas expresiones son equivalentes.
